```python
#輸入資訊
#account
#passwords
#山道猴子
#5

account = input('FB account/email:')
passwords = input('FB passwords:')
keywords = input('Keywords you want to collect:')
pages = input('how many pages you want to collect:')
```

```python
import time

#webdriver需要定時更新

path = r'chromedriver.exe'
service=Service(executable_path=path)

# 讓chromedriver透過selenium自動控制chrome瀏覽器

chrome=webdriver.Chrome(service=service)
chrome.get('https://mbasic.facebook.com/')

time.sleep(2)

#輸入帳號與密碼
input_xpath='/html/body/div/div/div[2]/div/table/tbody/tr/td/div[2]/div/div[2]/form/ul/li[1]/input'

element = chrome.find_element(By.XPATH, input_xpath)
element.send_keys(account)

input_xpath_2='/html/body/div/div/div[2]/div/table/tbody/tr/td/div[2]/div/div[2]/form/ul/li[2]/section/input'

element_2 = chrome.find_element(By.XPATH, input_xpath_2)
element_2.send_keys(passwords)


button_xpath = '/html/body/div/div/div[2]/div/table/tbody/tr/td/div[2]/div/div[2]/form/ul/li[3]/input'
button = chrome.find_element(By.XPATH,button_xpath)
button.click()

time.sleep(2)
#需要多一個步驟按下登入才會進到主頁
button_xpath_2 = '/html/body/div/div/div/div/table/tbody/tr/td/div/form/div/input'
button_2 = chrome.find_element(By.XPATH, button_xpath_2)
button_2.click()

#從主頁裡輸入關鍵字與按下搜尋鍵
time.sleep(2)

input_xpath_3='/html/body/div/div/div[1]/div/form/table/tbody/tr/td[2]/input'

element_3 = chrome.find_element(By.XPATH, input_xpath_3)
element_3.send_keys(keywords)

time.sleep(2)

button_xpath_3 = '/html/body/div/div/div[1]/div/form/table/tbody/tr/td[3]/input'
button_3 = chrome.find_element(By.XPATH,button_xpath_3)
button_3.click()

# 開始收集使用者的名稱
name = []

for i in range(int(pages)):
    print(f'-----collect page {i}-----')
    time.sleep(2)
    try:
        link_xpath = '/html/body/div/div/div[2]/div[2]/div[1]/div[2]/div[2]/a'
        link = chrome.find_element(By.XPATH, link_xpath)
        link.click()
        soup = BeautifulSoup(chrome.page_source, "lxml")
        bi = soup.find_all('h3',{'class':'ch ci cj ck'})
        print('user name:', end=' ')
        for n in bi:
            name.append(n.strong.text)
            print(n.strong.text, end=' ')
    except Exception as e: #當程式沒有再找到使用者時跳出迴圈
        print(e)
        break
    print(' ')
```

```python
import pandas as pd

d = {'使用者名稱':name
}

df = pd.DataFrame(d)
df
```

```python
# 用今天的日期來做為檔名

from datetime import datetime

date=datetime.now().strftime('%Y-%m-%d')  # m要用小寫字，因大本字是指分鐘

df.to_csv(f'FB_public_opinion_collection_{date}.csv',encoding='utf-8-sig')
```
