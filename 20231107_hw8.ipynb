```python
from tool import get_soup

url = 'https://www.imdb.com/chart/moviemeter/'
```

```python
s = get_soup(url)
mvs = s.find('script', id = '__NEXT_DATA__')
```

```python
#本週、上週排名、片名、連結(片長、上映年度、類型、rating、預告片連結、海報)
```

```python
import json

dmvs = json.loads(mvs.string)
type(dmvs)
```

```python
data = dmvs['props']['pageProps']['pageData']['chartTitles']['edges']
```

```python
name = []
year = []
link = []
ranking = []
lw_ranking = []
rating = []
genre = []
pic = []
leng = []
pre = []
for i in range(len(data)):
    try:
        name.append(data[i]['node']['originalTitleText']['text'])
        year.append(data[i]['node']['releaseYear']['year'])
        link.append('https://www.imdb.com/title/'+data[i]['node']['id'])
        rating.append(data[i]['node']['ratingsSummary']['aggregateRating'])
        genre.append(data[i]['node']['titleGenres']['genres'][0]['genre']['text'])
        
        if data[i]['node']['primaryImage'] == None:
            pic.append('None')
        else:
            pic.append(data[i]['node']['primaryImage']['url'])
            
        print(f'{i+1} finish1')
        
        if data[i]['node']['meterRanking']['rankChange']['changeDirection'] == 'DOWN':
            lw = data[i]['node']['meterRanking']['currentRank'] - data[i]['node']['meterRanking']['rankChange']['difference']

        elif data[i]['node']['meterRanking']['rankChange']['changeDirection'] == 'UP':
            lw = data[i]['node']['meterRanking']['currentRank'] + data[i]['node']['meterRanking']['rankChange']['difference']

        else:
            lw = data[i]['node']['meterRanking']['currentRank']
        ranking.append(data[i]['node']['meterRanking']['currentRank'])
        lw_ranking.append(lw)
        print(f'{i+1} finish2')
        
        subsoup=get_soup('https://www.imdb.com/title/tt15654328/?ref_=chtmvm_t_23')
        ul=subsoup.find('ul',class_='ipc-inline-list ipc-inline-list--show-dividers sc-afe43def-4 kdXikI baseAlt')
        le=ul.find_all('li',class_='ipc-inline-list__item')
        if len(le)==3:
            length=le[2].text.strip()
        else:
            length=le[1].text.strip()
        leng.append(length)

        prem=subsoup.find('a',class_='ipc-lockup-overlay ipc-focusable').get('href')
        pre.append('https://imdb.com'+prem)
        
        print(f'{i+1} finish3')
    
    except Exception as e:
        
        print(f'{i+1} fall',e)
        pass
    
    
```

```python
import pandas as pd
d = {'current ranking':ranking,
        'previous ranking':lw_ranking,
        'title':name,
        'link':link,
        'length':leng,
        'trailer':pre,
        'year':year,
        'IMDb rating':rating,
        'genre':genre,
        'poster':pic 
}

df = pd.DataFrame(d)
df
```

```python
# 用今天的日期來做為檔名

from datetime import datetime

date=datetime.now().strftime('%Y-%m-%d')  # m要用小寫字，因大本字是指分鐘
date

df.to_csv(f'IMDb電影排行榜_{date}.csv',encoding='utf-8-sig')
```
